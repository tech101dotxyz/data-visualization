Make the graph of first 5 countires on the basis of their total number of players. 

```python
fifa['nationality'].value_counts()[0:5].keys() # We are using keys to get the name of first 5 countries. 
```

```python
plt.figure(figsize=(8,5)) 
plt.bar(    list(fifa['nationality'].value_counts()[0:5].keys()),     list(fifa['nationality'].value_counts()[0:5]),      color="g")  # We converted the keys to list to use them for x-axis and the total number of games for y-axis. 
# Please explain that it is important to convert the data to list. 
```

```python
player_salary = fifa[['short_name','wage_eur']] # Let's find the name and wage of player. 
player_salary.head()
```

```python
player_salary=player_salary.sort_values(by=['wage_eur'],ascending=False) # First five. Here the ascending = False means that it will be in descending order. By default it is ascending. 
player_salary.head()
```

```python
# Let's make a graph of players and then their wages. 
```

```python
plt.figure(figsize=(8,5))
plt.bar(      list(player_salary['short_name'])[0:5], list(player_salary['wage_eur'])[0:5],  color='r') 
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color=["blue","green","red","black","pink"]) # Make it of different colours by using color parameter. 
plt.show()
```
