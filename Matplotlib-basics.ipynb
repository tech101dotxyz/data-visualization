We worked a lot on numbers and we also learnt how to find the releveant information from .csv file using Pandas. Now we will see how we can make graphs. We have all seen different types of graphs. Do you guys watch sports. You must have seen differnt types of graphs.  

Matplotlib is one of the most popular Python packages used for data visualization. It is a cross-platform library for making 2D plots from data in arrays.


```python
import matplotlib.pyplot as plt 
import numpy as np 
```

# That the numbers written here will make the y-axis. And the index position will show the x-axis. 


```python
plt.plot([1,2,3,4]) # The numbers written here will make the y-axis. And the index position will show the x-axis. 
```

# In the next case the numbers which are written first will make the x-axis and the numbers after that makes the y-axis.

```python
plt.plot([1,2,3,4],[5,6,7,8]) 
```

# We can also change the colour of line. 

```python
plt.plot([3,4,5,6],[2,3,4,5],color="red")
```

# We can also make a dash line. By writing dashdot in linestyle parameter. 

```python
plt.plot([1,2,3,4],linestyle="dashed")
```

### Insted of linestyle we can also write ls. The linestyle can also be changed by writing :,--,-,-.. Basically supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

```python
plt.plot([1,2,3,4],ls="dashdot", linewidth=6)
```

# Let's make a line graph to see how you have scored from the last 5 years in your class. 



```python
plt.plot([3,4,5,6,7],[85,95,87,90,92],color="red") # Put the classes first. Then final percentage later. 
# Mention the lowest value is considered at the starting value on y axis. In this case it's 85.  
```

```python
# A better way of writing it is this way. 
grades=np.array([3,4,5,6,7])
percentages = np.array([85,95,87,90,92])
plt.plot(grades,percentages,color="red")
```

```python
# A better way of writing it is this way. 
grades=np.array([3,4,5,6,7])
percentages = np.array([85,95,87,90,92])
plt.plot(grades,percentages,color="red")

plt.xlabel("Grades") # Let's put a label for these things. 
plt.ylabel("Percentages")
plt.title('Grade Vs Percentage')

```

Start Angle
As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.

The startangle parameter is defined with an angle in degrees, default angle is 0:

image.png

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.

The explode parameter, if specified, and not None, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed:



```python

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```
