### Ques:  Let's take any character and checkout his different skills on pie chart to find out which one his major skill and weak skills. 


```python
antmanskills = good[good['Name']=='Ant-Man'] 
antmanskills
```

```python
# For making pie chart we need values in form of list. So we'll convert them in list. 
ant_list=list(antmanskills.values) 

print(ant_list) # Here we can see that the list also have 'Ant-Man' and 'good' written. We also don't want total. We don't want these so we'll do list slicing.

ant_list = ant_list[0][2:8] # We sliced the ant_list and put the values back in ant_list
print(ant_list)
```

```python
labels = list(antmanskills.columns.values) # Here it will contain all the labels of antman. We will slice them too.  
labels_name=labels[2:8]
plt.pie(ant_list, labels=labels_name) # Here we have first put the list with numbers and then we have put the labels name. 
plt.show() 
```
